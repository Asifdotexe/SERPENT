```python
from serpent import serpentify
```

## Visualizing a Function

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

serpentify(factorial, title="Factorial Logic")
```

## Visualizing Code Strings

```python
code = """
for i in range(10):
    if i % 2 == 0:
        print(f"{i} is even")
    else:
        continue
"""
serpentify(code, title="Loop Logic", rankdir="LR")
```
